# WHAT IS PANDAS REPLACE

Values to the dataframe are replaced with other values dynamically.

```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\User\Downloads\book1.csv')
df
```

```python
df.replace(to_replace='CSE',value='ART')
```

```python
df.replace('A1','A2')
```

```python
df.replace('[A-Za-z]',0,regex = True)#replaces all the string values with 0
```

```python
df.replace({'Name':'[A-Za-z]'},0,regex=True)
```

```python
df
```

```python
df.replace('srack123',method= 'ffill')
```

```python
df.replace(0,method='bfill',limit=1)
```

```python
df
```

```python
df.replace(0,100,inplace = True)#doesnot makes the copy of the dataset instead modify the original dataset
df
```

What is Pandas fillna

fillna fills the NaN values with given values(input)

Synatax:Dataframe.fillna()

```python
import pandas as pd
```

```python
df=pd.read_csv(r'C:\Users\User\Downloads\another.csv')
df
```

```python
df.fillna(0)
```

```python
df.fillna(2)
```

```python
df.fillna({'Name':'none','branch':'-'})
```

```python
df.fillna(method='ffill')#fills the forward value in the NaN place
```

```python
df.fillna(method='pad')#same as ffill
```

```python
df.fillna(method='bfill')
```

```python
df.fillna(method='ffill',axis=0)
```

```python
df.fillna(method='ffill',axis=1)
```

```python
df.fillna(method='bfill',axis=1)
```

```python
df.fillna(0,limit= 2)
```

```python
df.fillna(method='ffill',limit= 2)
```

```python
df.fillna(5, inplace =True)#modify the dataset
df
```
